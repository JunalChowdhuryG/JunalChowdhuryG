```python
import pandas as pd
import random
import datetime
from datetime import datetime, timedelta
```

```python
num_productos = 10
num_categorias = 6

class Producto:
    def __init__(self, nombre, categoria, precio):
        self.nombre = nombre
        self.categoria = categoria
        self.precio = precio

def generar_lista_productos(num_productos):
    productos = []
    count = 0
    for j in range(num_categorias):
        for i in range(num_productos):
            count = count + 1
            producto_x = Producto(nombre=f"P_0{count}",categoria= f"{chr(ord('A') + j)}",precio=round(random.uniform(8, 20),2))
            productos.append(producto_x)       
    return productos

productos = generar_lista_productos(num_productos)

sucursales = ["Norte", "Sur", "Este", "Oeste"]

fecha_inicio = datetime(2021, 1, 1)
fecha_fin = datetime(2023, 12, 31)

data = []

for i in range((fecha_fin - fecha_inicio).days + 1):
    fecha_actual = fecha_inicio + timedelta(days=i)
    for j in sucursales:
        for h in productos:   
            fecha = fecha_actual
            prod = h.nombre
            cat = h.categoria
            precio_unitario = h.precio
            
            if(random.randint(0, 3)==0):
                cant = 0
            else:
                cant = round(random.randint(1,20)*(1 + fecha_actual.month/12) * (1 - fecha_actual.day/32), 0) 
            monto_dia = round((precio_unitario * cant),0)
            sucursal = j
            empleado = f"{chr(ord('A') + sucursales.index(j))}_{random.randint(1,8)}"
            data.append((fecha,prod,cat,precio_unitario,cant,monto_dia,sucursal,empleado))
df = pd.DataFrame(data, columns=["fecha","producto","categoria", "precio_unitario", "cantidad", "monto", "sucursal","empleado"])
df['cantidad'] = df['cantidad'].astype(int)
df['monto'] = df['monto'].astype(int)
df = df[df['cantidad'] >0]
df.to_csv('ventas.csv',index=False) 
```

```python
import pandas as pd
import numpy as np

# Definir el número de filas que deseas generar
num_filas = 100

# Generar datos aleatorios para cada columna
fechas = pd.to_datetime(pd.date_range(start='2023-01-01', end='2023-12-31', periods=num_filas))
id_productos = np.random.randint(1, 1000, size=num_filas)
nombres_productos = np.random.choice(['Producto A', 'Producto B', 'Producto C'], size=num_filas)
cantidades_vendidas = np.random.randint(1, 100, size=num_filas)
precios_unitarios = np.random.uniform(10, 100, size=num_filas).round(2)
total_ventas = cantidades_vendidas * precios_unitarios
id_clientes = np.random.randint(1, 500, size=num_filas)
nombres_clientes = ['Cliente_' + str(i) for i in range(1, num_filas+1)]
id_vendedores = np.random.randint(1, 10, size=num_filas)
nombres_vendedores = ['Vendedor_' + str(i) for i in range(1, num_filas+1)]
regiones = np.random.choice(['Norte', 'Sur', 'Este', 'Oeste'], size=num_filas)
canales_venta = np.random.choice(['En línea', 'En tienda', 'Por teléfono'], size=num_filas)

# Crear el DataFrame
data = {
    'Fecha': fechas,
    'ID_Producto': id_productos,
    'Nombre_Producto': nombres_productos,
    'Cantidad_Vendida': cantidades_vendidas,
    'Precio_Unitario': precios_unitarios,
    'Total_Venta': total_ventas,
    'ID_Cliente': id_clientes,
    'Nombre_Cliente': nombres_clientes,
    'ID_Vendedor': id_vendedores,
    'Nombre_Vendedor': nombres_vendedores,
    'Region': regiones,
    'Canal_Venta': canales_venta
}

df = pd.DataFrame(data)
df.to_csv
# Mostrar las primeras filas del DataFrame generado
df.head()
```
